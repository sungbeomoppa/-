```python
#범례지정
plt.title('Legend')
data1 = [2,4,6,8]
data2 = [8,6,4,2]
plt.plot(data1,label = 'asc')
plt.plot(data2,label = 'desc')
plt.legend()
plt.show()
```

```python
#x축 y축 이름지정
x = [2,4,6,8]
y = [1,3,5,7]
plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
d1 = [2,4,6,8]
d2 = [8,6,4,2]
plt.plot(d1,color ='r', label = 'dashdot',linestyle = 'dashdot')
plt.plot(d2,color = 'g', label = 'dotted',linestyle = 'dotted')
plt.legend()
plt.show()
```

```python
plt.title('title attached graph')
x = [10,20,30,40]
y = [1,3,5,7]
plt.plot(x,y,color = 'b',marker = '^',markersize = 10,linestyle = 'dotted',label = 'dotted')
plt.xlim(0,50)
plt.ylim(-5,15)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
filename = '/content/drive/MyDrive/2022-2 빅데이터/fine_dust.xlsx'
data = pd.read_excel(filename,index_col = 'area') #인덱스번호 자체를 area로 바꿔버림
data2019 = data[2019]
```

```python
plt.figure(figsize = (15,4)) #plt 크기
plt.plot(data2019, color ='b', marker = 'v',linestyle = 'dashed')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('nobody trust me')
plt.grid()#모눈종이
plt.show()
```

```python
plt.figure(figsize = (15,4))
for year in range(2016,2020):
  chartdata = data[year]
  plt.plot(chartdata, marker = 'o',label = year)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('justonelastdance')
plt.legend()
plt.grid()
plt.show()
#선그래프
```

```python
data2016 = data[2016]
plt.figure(figsize = (15,4))
plt.bar(data2016.index,data2016,color = 'g') #bar(x,y) 막대그래프 함수
plt.ylim(35,55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('fuckup')
plt.grid()
plt.show()
```
